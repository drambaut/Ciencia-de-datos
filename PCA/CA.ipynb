# ¿Qué es el Análisis de Correspondencia Simple?

El análisis de correspondencia simple es una técnica para analizar relaciones entre variables categóricas y crear perfiles basados en las proyecciones de las variables originales a las nuevas dimensiones que crea. Esto es útil, por ejemplo, al analizar y visualizar datos de encuestas.

El CA procesa una tabla de contingencia bidireccional que muestra la distribución de frecuencias entre dos variables. Representa la distribución de frecuencias en dimensiones numéricas, ortogonales. Basándonos en la proximidad a lo largo de las primeras de estas dimensiones, podemos explorar visualmente las asociaciones de los individuos y las categorías.

Podemos investigar, por ejemplo, si existe una relación entre el interés en la política y datos demográficos como la edad. Además, podemos interpretar una dimensión generada por el CA como una nueva dimensión sintética, como "estatus", que captura varias categorías que juntas contribuyen a un "estatus" alto o bajo.

```python
import pandas as pd
import prince
import seaborn as sns
```

```python
data = sns.load_dataset('titanic').dropna(subset=['age'])
```

```python
data.head()
```

```python
data_new = data[['sex','embarked','class','embark_town']]
```

```python
mca = prince.MCA(n_components=2)
mca.fit(data_new)
```

```python
mca.plot(data_new,show_column_labels=True)
```
