## Que es la regresión logistica ?

La regresión logística es un método estadístico para predecir clases binarias. La variable de resultado o variable objetivo es dicotómica por naturaleza. Dicotómico significa que solo hay dos clases posibles. Por ejemplo, se puede usar para problemas de detección de cáncer. Calcula la probabilidad de ocurrencia de un evento.

Es un caso especial de regresión lineal donde la variable objetivo es de naturaleza categórica. Utiliza un logaritmo de probabilidades como la variable dependiente. ## Regresión lineal vs Regresion logistica

La regresión lineal te proporciona un resultado continuo, pero la regresión logística proporciona un resultado constante. Un ejemplo de resultado continuo es el precio de una casa o el precio de una acción. Ejemplos de resultados discretos incluyen predecir si un paciente tiene cáncer o no, o predecir si un cliente abandonará la empresa. La regresión lineal se estima utilizando el método de Mínimos Cuadrados Ordinarios (OLS por sus siglas en inglés), mientras que la regresión logística se estima utilizando el método de Estimación de Máxima Verosimilitud (MLE por sus siglas en inglés).

![imagen](https://images.datacamp.com/image/upload/v1660054820/Regression_charts_b9de7355cf.png)

## Estimación de Máxima Verosimilitud Vs. Método de Mínimos Cuadrados

La Estimación de Máxima Verosimilitud (MLE por sus siglas en inglés) es un método de maximización de "verosimilitud", mientras que los Mínimos Cuadrados Ordinarios (OLS por sus siglas en inglés) son un método de aproximación que minimiza la distancia. Maximizar la función de verosimilitud determina los parámetros que son más probables para producir los datos observados. Desde un punto de vista estadístico, MLE establece la media y la varianza como parámetros al determinar los valores paramétricos específicos para un modelo dado. Este conjunto de parámetros puede usarse para predecir los datos necesarios en una distribución normal.

Las estimaciones de Mínimos Cuadrados Ordinarios se calculan ajustando una línea de regresión a los puntos de datos dados que tiene la suma mínima de las desviaciones cuadradas (error cuadrático mínimo). Ambos se utilizan para estimar los parámetros de un modelo de regresión lineal. MLE asume una función de masa de probabilidad conjunta, mientras que OLS no requiere ninguna suposición estocástica para minimizar la distancia.

## Función Sigmoide

La función sigmoide, también llamada función logística, proporciona una curva en forma de ‘S’ que puede tomar cualquier número de valor real y mapearlo en un valor entre 0 y 1. Si la curva se dirige hacia el infinito positivo, y predicho se convertirá en 1, y si la curva se dirige hacia el infinito negativo, y predicho se convertirá en 0. Si la salida de la función sigmoide es más de 0.5, podemos clasificar el resultado como 1 o SÍ, y si es menos de 0.5, podemos clasificarlo como 0 o NO. Por ejemplo: Si la salida es 0.75, podemos decir en términos de probabilidad: Hay un 75 por ciento de posibilidades de que un paciente padezca cáncer.

## Tipos de Regresión Logística

* Regresión Logística Binaria: La variable objetivo tiene solo dos posibles resultados, como Spam o No Spam, Cáncer o No Cáncer.

* Regresión Logística Multinomial: La variable objetivo tiene tres o más categorías nominales, como predecir el tipo de Vino.

* Regresión Logística Ordinal: La variable objetivo tiene tres o más categorías ordinales, como la calificación de un restaurante o producto de 1 a 5.

Vamos a construir el modelo de predicción de diabetes. Aquí, vas a predecir la diabetes utilizando el Clasificador de Regresión Logística.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv("diabetes.csv")
pima.columns = col_names
```

```python
pima.head()
```

```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = pima[feature_cols] # Features
y = pima.label # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)
```

```python
logreg = LogisticRegression(random_state=16)

logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
```

```python
y_test.describe()
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
cnf_matrix = np.array([[5, 2], [3, 5]])  # Aquí deberías poner tu matriz de confusión real
class_names = [0, 1]  # Nombres de las clases
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# Crear heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
target_names = ['without diabetes', 'with diabetes']
print(classification_report(y_test, y_pred, target_names=target_names))
```

La curva Característica Operativa del Receptor (ROC, por sus siglas en inglés) es un gráfico de la tasa de verdaderos positivos contra la tasa de falsos positivos. Muestra el equilibrio entre la sensibilidad y la especificidad.

y_pred_proba = logreg.predict_proba(X_test)[::,1]

Aquí, y_pred_proba está almacenando las probabilidades predichas de la clase positiva (asumida como 1) para cada muestra en el conjunto de prueba X_test, utilizando un modelo de regresión logística logreg.

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
