```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Mall_Customers.csv")
```

```python
X = data[['Age', 'Spending Score (1-100)']].copy()
```

```python
X.head()
```

```python
X.describe()
```

```python
result = []
eje_x = range(2,15)
for i in eje_x:
  kmeans_models = KMeans(n_clusters=i, n_init="auto")
  kmeans_models.fit(X)
  result.append(kmeans_models.inertia_)
```

```python
plt.plot(eje_x,result)
```

```python
kmeans_model = KMeans(n_clusters=4)
kmeans_result = kmeans_model.fit_predict(X)
kmeans_clusters = np.unique(kmeans_result)
for kmeans_cluster in kmeans_clusters:
    index = np.where(kmeans_result == kmeans_cluster)
    plt.scatter(X.loc[index:, 0], X.iloc[index:, 1])

plt.show()
```
